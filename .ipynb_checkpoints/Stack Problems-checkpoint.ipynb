## 155. Min Stack

* https://leetcode.com/problems/min-stack/

```javascript
/**
 * initialize your data structure here.
 */

/*
2 stack solution:

* essentially, you have 1 stack for all values
* then another stack for just min
    - if new_value <= currentMin, then we push it onto the min stack
* and when we pop from stack, we check if the top of min === stack, so we can pop it off too
* this allows for O(1) getMin()
*/
var MinStack = function() {
    this.stack = [];
    this.min = [];
};

/** 
 * @param {number} val
 * @return {void}
 */
MinStack.prototype.push = function(val) {
    if ( !Number.isNaN(val) ) {
        this.stack.push(val);
        if (this.min.length === 0 || val <= this.min[this.min.length - 1]) {
            this.min.push(val);
        }
    }
};

/**
 * @return {void}
 */
MinStack.prototype.pop = function() {
    const val = this.stack.pop();
    if (val === this.min[this.min.length - 1]) {
        this.min.pop();
    }
};

/**
 * @return {number}
 */
MinStack.prototype.top = function() {
    return this.stack[this.stack.length - 1];
};

/**
 * @return {number}
 */
MinStack.prototype.getMin = function() {
    return this.min[this.min.length - 1];
};

/** 
 * Your MinStack object will be instantiated and called as such:
 * var obj = new MinStack()
 * obj.push(val)
 * obj.pop()
 * var param_3 = obj.top()
 * var param_4 = obj.getMin()
 */
```

```javascript
/* 1 stack solution

* essentially, instead of pushing onto a min Stack, we push an object into the array
* object = {
    val: new value being pushed,
    min: current min value. if stack.length = 0, then min = val. else, Math.min(min, previous_value's min)
}

* you could probably do this with a linked list as well
*/
var MinStack = function() {
    this.stack = [];
};

/** 
 * @param {number} val
 * @return {void}
 */
MinStack.prototype.push = function(val) {
    if ( !Number.isNaN(val) ) {
        this.stack.push({
            val,
            min: this.stack.length === 0 ? val : Math.min(val, this.stack[this.stack.length - 1].min)
        });
    }
};

/**
 * @return {void}
 */
MinStack.prototype.pop = function() {
    this.stack.pop();
};

/**
 * @return {number}
 */
MinStack.prototype.top = function() {
    return this.stack[this.stack.length - 1].val;
};

/**
 * @return {number}
 */
MinStack.prototype.getMin = function() {
    return this.stack[this.stack.length - 1].min;
};
```
