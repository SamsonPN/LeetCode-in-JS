## 206. Reverse Linked List (easy):
* Reverse a singly linked list.
* Example:
    - Input: 1->2->3->4->5->NULL
    - Output: 5->4->3->2->1->NULL

***
Solution:
1. create 3 pointers: following, current, and previous
    - current = head;
    - following = head;
    - previous = null
2. while current !== null
    - following = current
    - current.next = previous
    - previous = current;
    - current = following
    - essentially, following is used to keep track of current.next and previous is used for current.previous
    - as we update current.next = previous, we can then use following to go to the next node until we reach the end of the list
    - previous = null is used in the case for an empty linked list, in which we can just return null
3. return previous if it asks you to return the head of the linked list since previous will be a non-null value

```javascript
// iterative
var reverseList = function(head) {
    let following = head;
    let current = head;
    let previous = null;
    
    while(current !== null) {
        following = current.next;
        current.next = previous;
        previous = current;
        current = following;
    }
    return previous;
};

// recursive
var reverseList = function(head) {
    const traverse = (following, current, previous) => {
        if(current == null) { return previous; }
        following = current.next;
        current.next = previous;
        previous = current;
        current = following;
        return traverse(following, current, previous);
    }
    
    return traverse(head, head, null);
}

// es6 syntax for in-place
var reverseList = function(head) {
    let [prev, current] = [null, head]
    while(current) {
        [current.next, prev, current] = [prev, current, current.next]
    }
    return prev
}
```

## 876. Middle of the Linked List (easy):
* Given a non-empty, singly linked list with head node head, return a middle node of linked list.
* If there are two middle nodes, return the second middle node.

* Example 1:
    - Input: [1,2,3,4,5]
    - Output: Node 3 from this list (Serialization: [3,4,5])
    - The returned node has value 3.  (The judge's serialization of this node is [3,4,5]).
    - Note that we returned a ListNode object ans, such that:
        * ans.val = 3
        * ans.next.val = 4
        * ans.next.next.val = 5
        * and ans.next.next.next = NULL.

* Example 2:
    - Input: [1,2,3,4,5,6]
    - Output: Node 4 from this list (Serialization: [4,5,6])
    - Since the list has two middle nodes with values 3 and 4, we return the second one.
 

* Note: The number of nodes in the given list will be between 1 and 100.

```javascript
// use of a slow pointer and a fast pointer
// once the fast pointer is at the end, slow pointer is at mid

var middleNode = function(head) {
    let slow = head;
    let fast = head;
    
    while(fast !== null && fast.next !== null) {
        slow = slow.next;
        fast = fast.next.next;
    }
    
    return slow;
};
```

## 83. Remove Duplicates from Sorted List

* https://leetcode.com/problems/remove-duplicates-from-sorted-list/

```javascript
/**
 * Definition for singly-linked list.
 * function ListNode(val, next) {
 *     this.val = (val===undefined ? 0 : val)
 *     this.next = (next===undefined ? null : next)
 * }
 */
/**
 * @param {ListNode} head
 * @return {ListNode}
 */

var deleteDuplicates = function(head) {
    if (head == null || head.next == null) return head;
    
    let prev = head;
    let current = head.next;
    
    while (current !== null) {
        if (current.val === prev.val) {
            prev.next = current.next;
            current = current.next;
            continue;
        }
        else {
            prev = current;
            current = current.next;
        }
    }
    
    return head;
};
```
